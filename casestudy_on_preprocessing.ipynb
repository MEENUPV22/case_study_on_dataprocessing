### Handling the missing values

```python
df=data.drop(columns=["No of Times Visited"])
df
```

```python
#handling missing values
num_df = df.select_dtypes(include="number")
cat_df = df.select_dtypes(include="object_")
num_col = num_df.columns.tolist()
cat_col = cat_df.columns.tolist()

print("Numerical columns are \n{} \nCategorical columns are \n{}".format(num_col,cat_col))
```

```python
for col in num_col:
    num_df[col] = num_df[col].fillna(num_df[col].median())
    
```

```python
for col in cat_col:
    cat_df[col] = cat_df[col].fillna(cat_df[col].mode()[0])
```

```python
data = pd.concat([num_df, cat_df], axis=1)
```

### Outlier Removal

```python
plt.boxplot(num_a)
plt.show()
```

```python
def remove_outliers(data,column_name):
    q1=data[column_name].quantile(0.25)
    q3=data[column_name].quantile(0.75)
    iqr=q3-q1
    upper_bound=q3+1.5*iqr
    lower_bound=q1-1.5*iqr
    data[column_name]=data[column_name].clip(upper=upper_bound)
    data[column_name]=data[column_name].clip(lower=lower_bound)
    return data[column_name]
```

```python
for col in num_cols1:
    num_a[col]=remove_outliers(num_a, col)
```
